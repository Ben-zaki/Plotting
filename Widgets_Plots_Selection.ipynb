```python
import hvplot.pandas
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import display
from ipywidgets import* 
```

```python
freq=5;T=1./freq
t=np.linspace(-T,T,2001)
pi=np.pi
```

```python
def plot(m,n,y,lw,color,ls,fig_name):
    current=np.cos(2*pi*freq*t+m*pi/3)
    voltage=np.cos(2*pi*freq*t+n*pi/3)
    inst_power=voltage*current
    signal_dict={'time':t,'voltage':voltage,
             'current':current,'inst_power':inst_power}
    df_signal=pd.DataFrame(signal_dict)
#------------------------------------------------------------------------------- 
    df_signal.plot.line(x='time',y=y,legend=True,grid=True,figsize=(12,4),
                        title='Current_Voltage_Power_plots',
                        xlim=(df_signal['time'].min(),df_signal['time'].max()),
                        ylim=(-1,1),fontsize=16,lw=lw,color=color,ls=ls)
    
    plt.xlabel('time(s)',fontsize=16)
    plt.ylabel('volt_curt_power',fontsize=16)
    #df_signal.to_excel('data_6.xlsx')
    #plt.savefig(fig_name, facecolor=facecolor, edgecolor='black',alpha=0.9)
    #display(df_signal[['time','voltage','current','inst_power']].head(3))
 
```

```python
y=[['voltage','current','inst_power'],['voltage','current'],
   ['voltage'],['current'],['inst_power']]
color=[['purple','orange','green'],['navy','chocolate','cyan'],
       ['brown','chartreuse','coral'],['grey','chocolate','fuchsia'],
       ['magenta','orange','sienna'],['#aa1155','#11ccff','gold']]
lw=['2.','3.','4','5.','6.','7.','8.']
n=(0,4,0.1)
m=(0,4,0.1)
ls=['-','--','-.']
fig_name=['fig.pdf','fig.png','fig.jpeg']
```

```python
my_plot=interactive(plot,y=y,color=color,lw=lw, n=n,
                        m=m,ls=ls,fig_name=fig_name)
#my_plot
```
